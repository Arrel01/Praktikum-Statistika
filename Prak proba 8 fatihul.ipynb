```python
import numpy as np
from scipy.stats import chi2_contingency

# Masukkan data tabel sesuai soal
data = np.array([
    [14, 6, 9],
    [10, 16, 10],
    [2, 13, 20]
])

# Uji chi-square
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square hitung:", round(chi2, 5))
print("df:", dof)
print("p-value:", p)
print("Tabel expected:")
print(np.round(expected, 2))
```
